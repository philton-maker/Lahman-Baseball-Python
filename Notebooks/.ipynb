### Question 1

### Imports and dataframes

```python
collegeplaying_df = pd.read_sql('SELECT * FROM collegeplaying;', con=engine)
collegeplaying_df.head()
people_df = pd.read_sql('SELECT * FROM people;', con=engine)
people_df.head()
salaries_df = pd.read_sql('SELECT * FROM salaries;', con=engine)
salaries_df.head()
vandy_df =collegeplaying_df[collegeplaying_df['schoolid'].str.contains('vandy')]
college_people_df =pd.merge(vandy_df,
         people_df,
         how = 'left',
         left_on = 'playerid',
         right_on = 'playerid')
```

### Find total Salary by player

```python
salaries_df = salaries_df.groupby('playerid')['salary'].sum()
```

```python
print(salaries_df)
```

### Merge Salaries with Vandy players in majors

```python
vandy_peopledf = pd.merge(college_people_df,
                          salaries_df,
                          how = 'left',
                          left_on = 'playerid',
                          right_on = 'playerid')
vandy_peopledf
```

### Sort and clean up

```python
vandy_salaries = vandy_peopledf[['namefirst', 'namelast', 'salary']]
```

```python
vandy_salaries = vandy_salaries.sort_values('salary', ascending=False)
```

```python
vandy_salaries.drop_duplicates()
```
